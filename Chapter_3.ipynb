### Construct the Overlap Graph of a collection of kmers

```python
from collections import defaultdict

def construct_overlap_graph(kmer_patterns):
    graph = defaultdict(list)
    
    for kmer1 in kmer_patterns:
        pref = kmer1[1:]
        for kmer2 in kmer_patterns:
            suf = kmer2[:-1]
            
            if pref == suf:
                graph[kmer1].append(kmer2)

    return '\n'.join([f"{key} -> {''.join(graph[key])}" for key in sorted(graph)])

kmer_patterns =  ['ATGCG','GCATG','CATGC','AGGCA','GGCAT']
# f = open('rosalind_ba3c.txt')
# kmer_patterns = [line for line in f.read().split('\n')]
# kmer_patterns = kmer_patterns[:-1]
# kmer_patterns[0].pop()
# print(len(kmer_patterns),kmer_patterns)
print(construct_overlap_graph(kmer_patterns))
     
```

## BA3F
### Find an Eulerian Cycle in a Graph

```python
# Solved
import random as rd
import numpy as np

def euler_cycle(graph, start = None):
    tour = []
    start = rd.choice([*graph.keys()]) if start is None else start
    
    def cycle(n):
        while len(graph[n]) > 0:
            nxt = graph[n].pop()
            cycle(nxt)
        tour.append(n)
    cycle(start)
    tour = tour[::-1]
    return '->'.join(tour)

def create_graph(inp):
    graph = defaultdict(list)
    for line in inp.split('\n'):
        start, end = line.split(' -> ')
        graph[start.strip()] = [j.strip() for j in end.split(',')]
    return graph

# graph = defaultdict(list)
inp = '''0 -> 3
         1 -> 0
         2 -> 1,6
         3 -> 2
         4 -> 2
         5 -> 4
         6 -> 5,8
         7 -> 9
         8 -> 7
         9 -> 6'''
f = open('rosalind_ba3f.txt')
inp = f.read().strip()
# print(inp)
graph = create_graph(inp)
# print(graph)
print(euler_cycle(graph))
```
