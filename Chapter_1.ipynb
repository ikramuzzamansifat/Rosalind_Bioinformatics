```python
from collections import defaultdict
from Bio import Seq

def Hamming_Distance(s1,s2):
    d = sum(i!=j for i,j in zip(s1,s2))
    return d

def ReversePattern(pattern):
    return Seq.reverse_complement(pattern)

def Reverse_Complement(text):
    t = len(text)
    d = {'A':'T', 'T':'A', 'G':'C', 'C':'C'}
    ans = []
    for i in text:
        ans.append(d[i])
    ans.reverse()
    return ''.join(ans)

def Neighbour(pattern, mismatch, words):
    if mismatch == 0:
        words.add(pattern)
    else:
        bases = ['A', 'T', 'G', 'C']
        p = len(pattern)
        
        for i in range(p):
            for j in range(4):
                new_pattern = pattern[:i] + bases[j] + pattern[i+1:]
#                 print(new_pattern)
                if mismatch <= 1:
                    words.add(new_pattern)
                else:
                    Neighbour(new_pattern, mismatch-1, words)
                
def MFP_with_mismatch_and_reverse_complement(text, k, d): # MFP -> Most Frequent Pattern
    allfrequentwords = defaultdict(int)
    l = len(text)
    
    for i in range(l-k+1):
        freqwords = set()
        Neighbour(text[i:i+k], d, freqwords)
        
        for word in freqwords:
            allfrequentwords[word] += 1
    
    for t in allfrequentwords.keys():
        reverse_k = Reverse_Complement(t)
        for i in range(l-k+1):
            if Hamming_Distance(text[i:i+k], reverse_k) <= d:
                allfrequentwords[t] += 1
    
    result = set()
#     print(max(allfrequentwords.values()))
    for t in allfrequentwords.keys():
        if allfrequentwords[t] == max(allfrequentwords.values()):
            result.add(t)
#             print(allfrequentwords[t])
            result.add(Reverse_Complement(t))
    
    return ' '.join(result)


# text = 'ACGTTGCATGTCGCATGATGCATGAGAGCT'
# k,d = 4,1
text = 'AAGTAAGCAAAGTAAGCACCGTACAAAAAGTAAGCACTACATACTACATACCGTACAAACCGTACAAACTGTAGTCTCCGTACAAACTACATAAGGATCCCAAGTAAGCACTGTAGTCTAAGTAAGCAAGGATCCCAAGTAAGCACCGTACAAACCGTACAAACTACATACTACATAAAGTAAGCACCGTACAAACCGTACAAAAAGTAAGCACCGTACAAAAGGATCCCCCGTACAAAAGGATCCCAAGTAAGCAAAGTAAGCACCGTACAAAAGGATCCCCTACATACCGTACAAAAGGATCCCCCGTACAAAAAGTAAGCACTGTAGTCTAGGATCCCCCGTACAAACTACATACCGTACAAAAGGATCCCCCGTACAAAAAGTAAGCACTGTAGTCTCTACATACTGTAGTCTCTGTAGTCTCCGTACAAACCGTACAAACTACATACTGTAGTCTCTGTAGTCTAAGTAAGCACTGTAGTCTCCGTACAAACCGTACAAAAGGATCCCCCGTACAAAAGGATCCCAGGATCCCCTGTAGTCTCTGTAGTCTCCGTACAAAAAGTAAGCACTACATACTACATACTACATACCGTACAAACTACATAAAGTAAGCACTACATACTGTAGTCTAAGTAAGCAAAGTAAGCACCGTACAAAAGGATCCCAGGATCCCAAGTAAGCACTGTAGTCTAAGTAAGCACCGTACAAACTACATACTACATAAAGTAAGCAAGGATCCCAGGATCCCAAGTAAGCACTGTAGTCTAAGTAAGCACTACATAAAGTAAGCACTGTAGTCTAAGTAAGCACTACATACCGTACAAA'
k,d = 5,2
print(MFP_with_mismatch_and_reverse_complement(text,k,d))

# s1 = 'GGGCCGTTGGT'
# s2 = 'GGACCGTTGAC'
# Hamming_Distance(s1,s2)
# # print(ReversePattern(s1))
# # print(Reverse_Complement(s1))
# print()
# words = set()
# print(words)
# Neighbour(s1,1, words)
# print(words)
```

```python
Symbol_to_number = {'A':0, 'C':1, 'G':2, 'T':3}

def Pattern_to_number(pattern):
    l = len(pattern)
    if(l == 0):
        return 0
    
    symbol = pattern[l-1]
    prefix = pattern[:l-1]
    return 4*Pattern_to_number(prefix) + Symbol_to_number[symbol]

# pattern = 'AGT'
pattern = 'GTAACCGGTGTCTTAAGTAG'
Pattern_to_number(pattern)
    
```

```python
Number_to_Symbol = {0: 'A', 1: 'C', 2: 'G', 3: 'T'}
def Number_to_Pattern(index, k):
    if k==1:
        return Number_to_Symbol[index]
    
    q = int(index/4)
    r = index%4
    return Number_to_Pattern(q,k-1) + Number_to_Symbol[r]

index,k = 45,4
# index,k = 7010,8
print(Number_to_Pattern(index,k))
    
```

```python
Symbol_to_digit = {'A': 0, 'C': 1, 'G': 2, 'T': 3}

def Pattern_to_number(pattern):
    l = len(pattern)
    if(l == 0): return 0
    
    return 4*Pattern_to_number(pattern[:l-1]) + Symbol_to_digit[pattern[l-1]]

def Frequency_Array(text, k):
    
    freq_arr = [0]*pow(4,k)
    
    l = len(text)
    for i in range(l-k+1):
        j = Pattern_to_number(text[i:i+k])
        freq_arr[j] += 1
    return freq_arr


# text = 'ACGCGGCTCTGAAA'
# k = 2
text = 'TAGTCTTACGCTGACTGCACTTTGTCTCCTTAATCAACCTCAGTGGCCTTCTTTGACGATCCTAATTTTAAGTAGATACCTGTAGACTGGTACGTGACGCCAACCCGACCGACTACAAACATAGTTCTACGCTGCGATACCAGAAATAGGGAGGCGCGGCTTAAATAGTCGACCACCCCTCGGGGGGGTCATTTTGTATGGAATCCGCTCACAGTCACGTCAACTAAATGAATATGTTCATCAGAGCGGTCAAGACCCGTCTTTCGTATACGTTAGTTGCCGTTGAGGAGCTTTAACTACTAGTCTACACGCCGAAGATCGTCCAAGCCGGCCTCATTTAACACGGCCGGCCAAGCGGTAGGAACAAATCACTAAGGCCCATCGTCGCACGGGGCATCCCGTCAGCGAATATTAGCGACCCGTGGAGCAGCTGCGTTCTTACTGTAAAGCTTCGGCCAGAGTCCGTGTACTCGTCGCTTTCATAATTAGGACGCACGGACTGTTCAACCTGTGATTCGGTGGCGGCCTCGTCTTTCCCAGAAGTTGCTCGAGAATCATATTGTTGGTGCGACTTATGAATGGTCATGTCTTGTGTGGTTCGGACACGAGTGGGAACCGTGTTACGACAGGGGCGTGGTATTGAGCGGACGTGATAGTGATTATAAAGCGATTGCGCCTTACACGATCATACAGACACCCGGGGAGATCCATCCTGATCAGAGAAAAAAAATTTTTTCGTCCTGAACTTTC'
k = 5
ans = Frequency_Array(text,k)

for i in ans:
    print(i, end = ' ')
```
